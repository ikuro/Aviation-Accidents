```python
def qty_band1(qqty):
    if(qqty<=5):
        return("5 and below")
    elif (qqty<=10):
        return("5 to 10")
    else:
         return("Over 10")
    
    
```

```python
df['qqty_band']=df['Quantity'].apply(qty_band1)
```

```python
df.head(3)
```

```python
conditions=[
    (df['Quantity']<=5),
    (df['Quantity']>5) & (df['Quantity']<=10),
    (df['Quantity']>10)
]
```

```python
choices=['5 and below','5 to 10', 'over 10']
```

```python
df['qty_band']=np.select(conditions, choices, default='Unknown')
```

```python
cities=['nairobi','mombasa','kisumu','thika','naivasha','nakuru']
```

```python
def proper_name():
    city_p=[]
    for i in cities:
        city_p.append(i.title())
    return city_p
```

```python
proper_name()
```

```python
def counterr():
    count=0
    for i in cities:
        count=count+1
    return count
```

```python
new_cities=['Lamu', 'wajir']
```

```python
def adder():
    new_addition=[new_cities]
    for i in cities:
        cities.extend(new_addition)

    return cities
```
